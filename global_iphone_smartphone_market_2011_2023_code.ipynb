```python
import numpy as np
import pandas as pd
import statistics as s
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iphone_ds['Percentage_of_iPhone_Users'] = iphone_ds['Percentage_of_iPhone_Users'].ffill()
iphone_ds['No_of_iPhone_Sold_USA'] = iphone_ds['No_of_iPhone_Sold_USA'].ffill()
iphone_ds['iOS_Market_Share'] = iphone_ds['iOS_Market_Share'].ffill()
iphone_ds['Android_Market_Share'] = iphone_ds['Android_Market_Share'].ffill()
```

```python
iphone_ds
```

```python
iphone_ds['per_iphone_user_usa'] = iphone_ds['No_of_iPhone_Users_USA']/iphone_ds['No_of_iPhone_Users'] * 100
```

```python
iphone_ds[['Year','No_of_iPhone_Users','No_of_iPhone_Users_USA']]
```

```python
plt.figure(figsize=(12, 6))
plt.plot(iphone_ds['Year'],iphone_ds['No_of_iPhone_Users'])
plt.title('Number of iPhone Users Globally')
plt.xlabel('Year')
plt.ylabel('Number of iPhone Users')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(iphone_ds['Year'], iphone_ds['No_of_iPhone_Users_USA'])
plt.title('Number of iPhone Users in USA')
plt.xlabel('Year')
plt.ylabel('Number of iPhone Users')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.title('Growth of Iphone users Globally vs USA')
plt.plot(iphone_ds['Year'],iphone_ds['No_of_iPhone_Users'], label = 'Global', marker = 'o')
plt.xlabel('Year')
plt.ylabel('Number of iPhone Users')
plt.plot(iphone_ds['Year'], iphone_ds['No_of_iPhone_Users_USA'], label = 'USA', marker = 'o')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(iphone_ds['Year'], iphone_ds['Percentage_of_iPhone_Users'], marker = 'o')
plt.title('Percentage of Iphone Users Globally')
plt.xlabel('Year')
plt.ylabel('Percentage Users')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(iphone_ds['Year'], iphone_ds['per_iphone_user_usa'], marker = 'o')
plt.title('Percentage of Iphone Users in USA')
plt.xlabel('Year')
plt.ylabel('Percentage Users')
plt.show()
```

```python
iphone_ds[['Year','No_of_iPhone_Sold', 'No_of_iPhone_Sold_USA', 'Percentage_of_iPhone_Users']]
```

```python
plt.figure(figsize=(12, 6))
plt.plot(iphone_ds['Year'], iphone_ds['No_of_iPhone_Sold_USA'], marker = 'o')
plt.title('Number of iphones sold in USA')
plt.xlabel('Year')
plt.ylabel('Number')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(iphone_ds['No_of_iPhone_Sold'], iphone_ds['Percentage_of_iPhone_Users'])
plt.title('Relationship b/w iphone sold vs Percentage Users')
plt.show()
```

```python
iphone_ds[['Year','iOS_Market_Share','Android_Market_Share']]
```

```python
plt.figure(figsize=(12, 6))
plt.plot(iphone_ds['Year'], iphone_ds['iOS_Market_Share'], label='iOS Market Share', marker='o')
plt.plot(iphone_ds['Year'], iphone_ds['Android_Market_Share'], label='Android Market Share', marker='o')
plt.xlabel('Year')
plt.ylabel('Market Share (%)')
plt.title('iOS vs Android Market Share Over Years')
plt.legend()
plt.show()
```

```python
bar_width = 0.4
years = np.arange(len(iphone_ds['Year']))

plt.figure(figsize=(12, 6))

plt.bar(years, iphone_ds['iOS_Market_Share'], width=bar_width, label='iOS Market Share')
plt.bar(years + bar_width, iphone_ds['Android_Market_Share'], width=bar_width, label='Android Market Share')

plt.xlabel('Year')
plt.ylabel('Market Share (%)')
plt.title('iOS vs Android Market Share Over Years')
plt.xticks(years + bar_width / 2, iphone_ds['Year'])
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
iphone_ds['Growth_Rate_Global'] = iphone_ds['No_of_iPhone_Users'].pct_change() * 100
iphone_ds['Growth_Rate_USA'] = iphone_ds['No_of_iPhone_Users_USA'].pct_change() * 100

plt.plot(iphone_ds['Year'], iphone_ds['Growth_Rate_Global'], label='Global Growth Rate', marker='o')
plt.plot(iphone_ds['Year'], iphone_ds['Growth_Rate_USA'], label='USA Growth Rate', marker='o')

plt.xlabel('Year')
plt.ylabel('Growth Rate (%)')
plt.title('Annual Growth Rate of iPhone Users')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
correlation_matrix = iphone_ds[['No_of_iPhone_Users', 'No_of_iPhone_Users_USA', 'Percentage_of_iPhone_Users',
                                'No_of_iPhone_Sold', 'No_of_iPhone_Sold_USA', 'iOS_Market_Share',
                                'Android_Market_Share', 'per_iphone_user_usa']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of iPhone Metrics')
plt.show()
```
